```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train = pd.read_csv('train.csv')
train.shape, train.info(), train.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter relevant numerical columns for correlation
numerical_cols = train.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = train[numerical_cols].corr()

# Plot heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# Loan Type vs Target
plt.figure(figsize=(10, 6))
sns.barplot(data=train, x="loan_type", y="target")
plt.title("Default Rates by Loan Type", fontsize=14)
plt.ylabel("Average Default Rate")
plt.xticks(rotation=45)
plt.show()

# New or Repeat Loans vs Target
plt.figure(figsize=(6, 4))
sns.barplot(data=train, x="New_versus_Repeat", y="target")
plt.title("Default Rates: New vs Repeat Loans", fontsize=14)
plt.ylabel("Average Default Rate")
plt.show()


```

##### In the above plots, we can see that those who took out loans of type 14, 2, 23, 20, and 15 have higher default rates. This can be a result of high interest rates. Also, new loan customers will most likely default compared to repeat loan customers. This can be due to lack of credit or bad risk assessments from lenders. ALso, repeat customers have good records because they naturally wouldnt apply for new loans knowing they defaulted their last

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=train, x="loan_type", y="target", hue="New_versus_Repeat", palette="husl")
plt.title("Default Rate by Loan Type and New vs Repeat Loans", fontsize=14)
plt.ylabel("Average Default Rate")
plt.xticks(rotation=45)
plt.legend(title="Loan Type")
plt.show()
```

##### The plot above confirms our early discovery that new loans are defaulted more often than repeat loans, with few exceptions

```python
# Countplot of Loan Types


unique_loantypes= train['loan_type'].value_counts()
print(unique_loantypes)


plt.figure(figsize=(12, 6))
sns.countplot(data=train, x='loan_type', hue='target', palette='husl')  # Removed palette
plt.title('Count of Each Loan Type', fontsize=14)
plt.xlabel('Loan Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()


# Countplot for New vs Repeat Loans
plt.figure(figsize=(6, 4))
sns.countplot(data=train, x="New_versus_Repeat", hue='target', palette='husl')
plt.title("New vs Repeat Loan Distribution", fontsize=14)
plt.show()
```

##### In the above plots, we can see that more people took out loan type 1. A major factor for this can be favourable loan terms like interest rates and repayment period. This also explains why most loan in type_1 are repeat customers. Nobody hates a considerate lender. Despite that, loan type 1 recorded few defaults probably due to very good risk assessment on the institutions' part and good credit record of customers

```python
# Convert dates to datetime format
train['disbursement_date'] = pd.to_datetime(train['disbursement_date'])
train['due_date'] = pd.to_datetime(train['due_date'])

# Aggregate default rates by month
monthly_defaults = train.groupby(train['disbursement_date'].dt.to_period("M"))['target'].mean().reset_index()
monthly_defaults['disbursement_date'] = monthly_defaults['disbursement_date'].dt.to_timestamp()

# Line plot of monthly default rates
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_defaults, x="disbursement_date", y="target", marker="o", color="blue")
plt.title("Monthly Default Rates Over Time", fontsize=14)
plt.xlabel("Disbursement Date")
plt.ylabel("Default Rate")
plt.grid(True, alpha=0.4)
plt.show()
```

```python
# Extract month from disbursement_date
train['disbursement_month'] = train['disbursement_date'].dt.month

# Barplot for default rates by month
plt.figure(figsize=(10, 6))
sns.barplot(data=train, x="disbursement_month", y="target", hue="New_versus_Repeat", palette="husl" )
plt.title("Default Rates by Disbursement Month", fontsize=14)
plt.xlabel("Disbursement Month")
plt.ylabel("Default Rate")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

##### In the above plots, loans disbursed between November and March will most likely be defaulted. There are different factors that might explain this. For example,  School fees(the school year in Kenya starts in January and parents might be struggling with their finances), increased spending during Christmas and new year (Kenyans like most Africans love to both Xmas and New Year and tend to spend a lot leaving them with less cash)

```python


# Plot total amount loaned out each month
plt.figure(figsize=(12, 6))
sns.barplot(data=train, x='disbursement_month', y='Total_Amount')
plt.title('Total Loan Amount Disbursed Each Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total Loan Amount')
plt.xticks(rotation=45)
plt.grid(alpha=0.4)
plt.show()
```

```python
# Extract month from disbursement_date
train['disbursement_month'] = train['disbursement_date'].dt.month
# Barplot for default rates by month
plt.figure(figsize=(10, 6))
sns.barplot(data=train, x="disbursement_month", y="Total_Amount" )
plt.title("Total Amount by Disbursement Month", fontsize=14)
plt.xlabel("Disbursement Month")
plt.ylabel("Total Amount")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# there are lot of high correlation reports e.g lender_id, total_amount
#to reduce number of high correlation variables and number of columns needed for training, i will create ratios 

train = train.drop(['country_id'], axis=1)
train['repayment_ratio'] = train['Total_Amount_to_Repay'] / train['Total_Amount']
train['lender_funded_ratio'] = train['Amount_Funded_By_Lender'] / train['Total_Amount']
train['lender_repaid_ratio'] = train['Lender_portion_to_be_repaid'] / train['Lender_portion_Funded']


#To simplify my data , I will drop disbursement and due dates, since i have duration already 
# and other redundant columns

train = train.drop(['ID', 'customer_id', 'tbl_loan_id', 'lender_id', 'disbursement_date', 'due_date',
                   'Amount_Funded_By_Lender', 'Lender_portion_Funded', 'Lender_portion_to_be_repaid',
                   'Total_Amount_to_Repay'], axis=1)
train.head()
```

```python
train.info()
```

```python
train['lender_repaid_ratio'] = train['lender_repaid_ratio'].fillna(train['lender_repaid_ratio'].mean())
train['lender_funded_ratio'] = train['lender_funded_ratio'].fillna(train['lender_funded_ratio'].mean())

numeric_cols = ['Total_Amount', 'duration', 'lender_repaid_ratio', 'lender_funded_ratio']
for col in numeric_cols:
    train[f'{col}_transformed'] = np.log1p(train[col])

train.head()
```

```python
train = train.drop(['Total_Amount', 'duration', 'lender_repaid_ratio', 'lender_funded_ratio'], axis=1)
train.head()
```

```python
#Encode categorical features (one-hot encoding)
train['New_versus_Repeat'] = train['New_versus_Repeat'].apply(lambda x: 1 if x == 'Repeat Loan' else 0)
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

train['loan_type'] = encoder.fit_transform(train['loan_type'])
train.head(50)
train.to_csv('../clean_train.csv', index=False)
```

### TRAINING 

```python
from imblearn.over_sampling import SMOTE

X = train.drop('target', axis=1)
y = train['target']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#scaler = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42, stratify=y_resampled)

# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

log_reg = LogisticRegression(max_iter=1000, random_state=42 )

# Train the model
log_reg.fit(X_train, y_train)

# Make predictions
y_pred = log_reg.predict(X_test)
y_probs = log_reg.predict_proba(X_test)[:, 1] 
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

roc_auc = roc_auc_score(y_test, y_probs)
print(f"\nROC AUC Score: {roc_auc:.4f}")
```
